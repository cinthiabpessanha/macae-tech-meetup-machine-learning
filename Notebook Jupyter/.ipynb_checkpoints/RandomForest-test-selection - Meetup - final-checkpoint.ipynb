```python
#import important libs - for data manipulation and math
import numpy as np
import pandas as pd

#testing the folder content where data is
import os
print(os.listdir("..\Datasets")) #directory where data is
```

```python
#creating variables to load data from CSV files - pandas helps a lot
folder = "..\Datasets"

pokemon_df = pd.read_csv(folder + "\\pokemon.csv")
combats_df = pd.read_csv(folder + "\\combats.csv")
tests_df = pd.read_csv(folder + "\\tests.csv")

pokemon_df = pokemon_df.fillna('NA')
#just to see data organization
print(pokemon_df)
```

```python
merged1 = combats_df.merge(pokemon_df, how = "left", left_on = "First_pokemon", right_on = "#")
merged2 = combats_df.merge(pokemon_df, how = "left", left_on = "Second_pokemon", right_on = "#")

merged2.drop(["Winner","First_pokemon","Second_pokemon", "Generation", "Name"], axis = 1, inplace = True)
merged1.drop(["Generation", "Name"], axis = 1, inplace = True)

for i in merged2.columns : 
    merged2.rename(columns = {i : i + "_2"}, inplace = True)

data_train = pd.concat([merged1,merged2], axis = 1)

#the result of the merges
data_train.sample(5)
```

```python
for col in data_train.columns :       
    if (data_train[col].dtype == "bool") :
        data_train[col] = data_train[col].astype(int)  # Just change bool to int
        
data_train['Type1']=data_train['Type1'].astype('category')
data_train['Type2']=data_train['Type2'].astype('category')
data_train['Type1_2']=data_train['Type1_2'].astype('category')
data_train['Type2_2']=data_train['Type2_2'].astype('category')

# This feature below will be our label/target
data_train["Winner_2"] = (data_train["Winner"] == data_train["First_pokemon"]).astype(int)

X = data_train.drop(["Winner","Winner_2", "First_pokemon", "Second_pokemon", "#", "#_2"], axis = 1)
y = data_train.Winner_2

X.head()
```

```python
from sklearn.preprocessing import OneHotEncoder



encoder = OneHotEncoder(categories = "auto", handle_unknown = "ignore")
X_train_encoded = encoder.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(X_train_encoded, y, test_size=0.35, random_state=120)
```

```python
clf = RandomForestClassifier(n_estimators=110, n_jobs=-1 )
model = clf.fit(x_train, y_train) 
pred = model.predict(x_test)
print('Accuracy of ', accuracy_score(pred, y_test)*100)



```

```python
from sklearn.model_selection import cross_val_score


# Cross validation with 10 folds
scores = cross_val_score(clf, x_test, y_test, cv = 10)
print(scores) 
print('Standard dev.', scores.std())
```

```python
feats = {} # a dict to hold feature_name: feature_importance

for feature, importance in zip(X.columns, model.feature_importances_):
    feats[feature] = importance #add the name/value pair 

feats_sum = {}
for feature in feats:
    if('_2' in feature):
        gen_feature = feature.translate({ord(i): None for i in '_2'})
        if(gen_feature in feats_sum):
            feats_sum[gen_feature] = feats_sum[gen_feature] + feats[gen_feature]
        else:
            continue
    else:
        feats_sum[feature] = feats[feature]
    
importances = pd.DataFrame.from_dict(feats_sum, orient='index').rename(columns={0: 'Importance'})
importances.sort_values(by='Importance').plot(kind='bar', rot=45)
```

```python
from sklearn.externals import joblib

# save the model to disk
filename = 'finalized_model.sav'
joblib.dump(model, filename)

```

```python
#save the encoder to disk
encoderFileName = 'encoder.sav'
joblib.dump(encoder, encoderFileName)
```

```python
#loading model from disk - https://machinelearningmastery.com/save-load-machine-learning-models-python-scikit-learn/

loaded_model = joblib.load(filename)

loaded_encoder = joblib.load(encoderFileName)
```

```python
folder = "..\Datasets"
#teste-individual.csv contains one single row to test the ML model and a header with the features names 
tests_individuals_df = pd.read_csv(folder + "\\test-individual.csv")

tests_individuals_df = tests_individuals_df.fillna('NA')

for col in tests_individuals_df.columns :
    if (tests_individuals_df[col].dtype == "bool") :
        tests_individuals_df[col] = tests_individuals_df[col].astype(int)  # Just change bool to int
        
print(tests_individuals_df.sample(1))

print(loaded_model.predict(loaded_encoder.transform(tests_individuals_df)))

```
